# My model

```python
def initialisation(X):
    W= np.random.randn(X.shape[1], 1)
    b= np.random.randn(1)
    return (W,b)

def model(X, W, b):
    Z= X.dot(W)+b
    A= 1/(1+np.exp(-Z))
    return A

def log_loss(A, y):
    epsilon = 1e-15
    return (1/len(y))*np.sum(-y * np.log(A+epsilon)-(1-y)*np.log(1-A+epsilon))

def gradients(A, X, y):
    dW = (1/len(y))*np.dot(X.T, A-y)
    db = (1/len(y))*np.sum(A-y)
    return dW, db

def update(dW, db, W, b, learning_rate):
    W = W - learning_rate*dW
    b = b - learning_rate*db
    return (W, b)

def predict(X, W, b):
    A = model(X, W, b)
    print(A)
    return A >=0.5

from sklearn.metrics import accuracy_score
from tqdm import tqdm

def artificial_neuron(X, y, learning_rate=0.1, n_iterations=100):
    #initialiser W et b
    W, b = initialisation(X)
    
    Loss=[]
    acc = []
    
    for i in tqdm(range(n_iterations)):
        A = model(X, W, b)
        if i %150 ==0:
            Loss.append(log_loss(A, y))
            acc.append(accuracy_score(y,y_pred=predict(X,W,b)))
            
        dW, db = gradients(A, X, y)
        W, b = update(dW, db, W, b, learning_rate)
        
    #y_pred = predict(X, W, b)  
    #print(accuracy_score(y,y_pred))
    
    plt.figure(figsize=(24,10))
    plt.subplot(1,2,1)
    plt.plot(acc)
    plt.title('accuracy')
    
    plt.subplot(1,2,2)
    plt.plot(Loss)
    plt.title('loss function - fonction coût')
    plt.show()
    
    return (W,b)

def artificial_neuron_2(X_train, y_train, X_test, y_test, learning_rate=0.1, n_iterations=100):
    #initialiser W et b
    W, b = initialisation(X)
    
    train_loss=[]
    train_acc = []
    
    test_loss = []
    test_acc = []
    
    for i in tqdm(range(n_iterations)):
        A = model(X, W, b)
        y_pred=predict(X_train,W,b))
        if i %150 ==0:
            train_loss.append(log_loss(A, y_train))
            train_acc.append(accuracy_score(y_train, y_pred))
            
        dW, db = gradients(A, X_train, y_train)
        W, b = update(dW, db, W, b, learning_rate)
        
    #y_pred = predict(X, W, b)  
    #print(accuracy_score(y,y_pred))
    
    plt.figure(figsize=(24,10))
    plt.subplot(1,2,1)
    plt.plot(acc)
    plt.title('accuracy')
    
    plt.subplot(1,2,2)
    plt.plot(Loss)
    plt.title('loss function - fonction coût')
    plt.show()
    
    return (W,b)
```
