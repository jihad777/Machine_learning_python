## Line plot

```python
a = np.linspace(0,10,1000)
```

```python
sin_a = np.sin(a)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(a, sin_a)
plt.show()
```

```python
b = np.linspace(0,10,1000)
```

```python
sin_b = np.sin(b)
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(b, sin_b)
plt.xlabel('range from 1 to 10')
plt.ylabel('sinus from 1 to 10')
plt.title('Function')
plt.xlim(right=6)               # xlimite

plt.subplot(1,2,2)
plt.plot(b, sin_b)
plt.xlabel('range from 1 to 10')
plt.ylabel('sinus from 1 to 10')
plt.title('Function')
```

```python
plt.figure()
plt.plot(a, np.sin(a), color='g')
plt.plot(a, np.cos(a), color='r', linestyle= '--')
plt.show()
```

```python
plt.figure()
plt.plot(a, np.sin(a), 'k:')    #colors and linestyle can be combined into a single non-keyword argument
plt.plot(a, np.cos(a), 'r--')
plt.show()
```
