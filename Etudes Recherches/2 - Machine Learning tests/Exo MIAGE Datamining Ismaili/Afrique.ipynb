```python
from sklearn.preprocessing import StandardScaler
```

```python
afrique_sans_pays = afrique.drop('Pays', axis = 1)
```

```python
# Normalisation >> (Standardisation)

scaled_ft = StandardScaler().fit_transform(afrique_sans_pays)
afrique_scaled = pd.DataFrame(scaled_ft, index= afrique_sans_pays.index, columns=afrique_sans_pays.columns)
```

```python
afrique_scaled.head()
```

```python
# éliminer les valeurs nulles
df = afrique_scaled.dropna(axis=0)
```

# Méthode du coude:

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)  #l'inertie
plt.figure(figsize=(16,8))
plt.plot(range(1,11),wcss, 'bx-')
plt.title('Methode du coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('WCSS')
plt.show()
```

# Etant donné qu'il n'est pas clair de la méthode du coude le numbre K idéal, nous allons essayer le silhouette_score

### silhouette_score

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_sc = []
for i in range(2,10):
    model = KMeans(n_clusters=i, random_state=0)
    c = model.fit_transform(df)
    silhouette_sc.append(silhouette_score(df, model.fit_predict(df)))
plt.figure()
plt.bar(range(2,10), silhouette_sc)
plt.title('Silhouette score')
plt.show()
```

# Modèle

```python
# Entrainement du modèle en utilisant le meilleur K, 2 dans cas 

kmeans = KMeans(n_clusters=2, random_state=0).fit(df)

pred = kmeans.predict(df)

# Ajouter le numéro de cluster au jeu de données initial

afrique['cluster'] = pred +1
```

```python
afrique.sort_values(by='cluster', ascending=False).round(decimals=2)
```

```python
afrique.sort_values(by='cluster', ascending=False).round(decimals=2).groupby(by='cluster').mean().round(decimals=2)

# Moyenne du cluster par attribut
```
