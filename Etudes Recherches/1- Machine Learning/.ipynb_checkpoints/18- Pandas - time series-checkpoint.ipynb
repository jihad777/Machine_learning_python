```python
bitcoin['2021']['Close'].resample('M').std().plot()
```

```python
#Graph sympas
m = bitcoin['Close']['2021'].resample("W").agg(['mean', 'std', 'min', 'max'])

plt.figure(figsize=(12,8))
m['mean']['2021'].plot(label= 'moyenne de fermeture par semaine')
plt.fill_between(m.index, m['max'], m['min'], alpha= 0.2, label='min-max par semaine')

plt.legend()
plt.show()
```

```python
#Moving average
plt.figure(figsize=(12,8))
bitcoin.loc['2021-08']['Close'].rolling(window= 7).mean().plot(ls=':')
bitcoin.loc['2021-08']['Close'].plot(lw=0.5, ls="-", c='r')
bitcoin.loc['2021-08']['Close'].ewm(alpha=0.6).mean().plot(ls=':', c="orange") #exponnetial weighted average: (moyenne mobile exponnentielle)
```

```python
#------------------------Exercice: technique de la tortue en trading-------------------------
plt.figure(figsize=(12,8))
bitcoin.loc['2021']['Close'].rolling(window= 28).max().plot(label = 'max', c='b')
bitcoin.loc['2021']['Close'].rolling(window= 28).min().plot(label = 'min', c='r')
bitcoin.loc['2021']['Close'].plot(label = 'Close')
plt.legend()
```

```python
bitcoin['Buy'] = np.zeros(len(bitcoin))
bitcoin['sell'] = np.zeros(len(bitcoin))
```

```python
bitcoin.head()
```

```python
for i in range (len(bitcoin)):
    if bitcoin['Close'] > bitcoin.loc['2021']['Close'].rolling(window= 28).max():
        bitcoin['Buy'] = 1
    elif bitcoin['Close'] < bitcoin.loc['2021']['Close'].rolling(window= 28).min():
        bitcoin['sell'] = 1
```
