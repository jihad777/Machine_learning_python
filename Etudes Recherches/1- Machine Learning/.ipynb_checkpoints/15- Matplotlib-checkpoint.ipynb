```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris 
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')
```

```python
#Nuage de points personnalis√©s
plt.scatter(x[:,0], x[:,1], c=y, alpha=1.5, s=x[:,2]*100) #c=y pour colorier les diff points du nuage selon la classe de y
                                                          #alpha = contraste
                                                          #s = taille des points, ici selon la troisi√®me variable de notre DTset
plt.xlabel('longeur s√©pal')
plt.ylabel('largeur s√©pal')
```

```python
#-----------Graphique 3D---------------
#pour pouvoir zoomer/d√©zoomer retourner le graph 3D 

from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')
ax.scatter(x[:,0],x[:,1], x[:,2], c=y )
```

```python
#--------------Histogramme-------------
#Extr√™mement utile en stats et ML pour savoir la distribution d'une variable et si elle suit une loi normale
plt.hist(x[:,0], bins=30)
plt.hist(x[:,1], bins= 2)
```

```python
#histogramme 2D
plt.hist2d(x[:,0],x[:,1], cmap='Blues')
plt.xlabel('longeur s√©pal')
plt.ylabel('largeur s√©pal')
plt.colorbar()  #color bar pour afficher la fr√©quence
```

```python
#---------------Contour plot----------------- ML = r√©soudre des probl√®mes d'optimisation 
#                                             (par ex: minimiser la fonction cout, maximiser les revenus....)
f = lambda x,y : np.sin(x)+ np.cos(x+y)
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)
Z = f(X,Y)
```

```python
plt.contour(X,Y,Z, 20, colors='black') #ligne continues = valeurs positives (max) / tir√©s = valeurs n√©gatives (min)
```

```python
plt.contourf(X,Y,Z,20,cmap='Blues')
plt.colorbar()
#avec √ßa on peut comprendre quelles sont les valeurs de notre gradient sur notre contour-plot
```
