```python
#------------Feature Ingeneering !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
data.loc[(data['age'] <= 20), 'age'] = 1
data.loc[(data['age'] > 20) & (data['age'] <= 30), 'age']= 2
data.loc[(data['age'] > 30) & (data['age'] <= 40), 'age']= 3
data.loc[data['age']>40, 'age'] = 4
```

```python
def categorie_age(age):
    if age <= 20:
        return 'age < 20'
    elif (age > 20) & (age <= 30):
        return '20 < age < 30'
    elif (age > 30) & (age <= 40):
        return '30 < age < 40'
    else:
        return 'age > 40'
```

```python
data['age'].map(categorie_age).value_counts()
```
