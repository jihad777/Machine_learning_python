# Preprocessing

```python
def encodage(df):
    dictionnaire_codes = {   'negative':0,
                             'positive':1,
                             'not_detected':0,
                             'detected':1 }
    
    for col in df.select_dtypes(object):
        df[col] = df[col].map(dictionnaire_codes)
        
    return df    
```

```python
def imputation(df):
    df = df.dropna(axis=0)
    return df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
to_scale_columns = ['Patient age quantile','Platelets', 'Red blood Cells', 'Leukocytes', 'Eosinophils','Monocytes']
```

```python
def scalling(df):
    scaller = StandardScaler()
    for col in to_scale_columns:
        df[col] = scaller.fit_transform(df[[col]])
     
    return df
```

```python
def preprocessing(df):
    df = scalling(df)
    df = encodage(df)
    df = imputation(df)
    
    X = df.drop('SARS-Cov-2 exam result', axis = 1)
    y = df['SARS-Cov-2 exam result']
    
    print(X.shape)
    print(y.shape)
    
    print(y.value_counts(normalize=True))
    
    return X, y
```

```python
X_train, y_train = preprocessing(trainset)
```

```python
X_test, y_test = preprocessing(testset)
```

```python
def evaluation(model):
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    
    N, train_score, test_score = learning_curve(model, X_train, y_train, cv=4, scoring='f1', train_sizes=np.linspace(0.1,1,50))
    
    
    plt.figure(figsize=(8,6))
    plt.plot(N, train_score.mean(axis=1), label=model)
    plt.plot(N, test_score.mean(axis=1), label='validation')
    plt.legend()
```

```python
for name, model in dictio_models.items():
    print(name)
    evaluation(model)
```

# Choix du seuil de décision

## Seuil de décision

### SVM -0,6
### LR -2,3

```python
def model_final(model, X, threshold):
    return model.decision_function(X) > threshold
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score
```

```python
def eval_model_manual_threshold(model,X, threshold):
    y_pred = model_final(model, X_test, threshold)
    
    print('f1-score: ', f1_score(y_test, y_pred))
    print('recall: ', recall_score(y_test, y_pred))
    print('precision: ', precision_score(y_test, y_pred))
```

```python
eval_model_manual_threshold(svm_normal, X_test, threshold=-0.6)
```

```python
eval_model_manual_threshold(grid_logistic, X_test, threshold=-2.3)
```
