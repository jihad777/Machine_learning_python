```python
plt.figure(figsize=(12,8))
plt.plot(x, y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('Debut de figure, contenu, fin?')
plt.xlabel('axe des abssices')
plt.ylabel('axe des ordonnees')
plt.legend()
plt.show()
plt.savefig('fig1.png')
```

```python
#Plusieurs graphs sur la mm figure
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
plt.plot(x, y)
plt.subplot(2,1,2)
plt.plot(x,x**7, c='r')
```

```python
#Méthode orientée objet: pour faire des graphs plus poussés (1% de plus de fonctionnalité que les méthodes du haut)
#par exemple afficher 2 graphes avec le mm abssice
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x,y)
ax[1].plot(x, np.sin(x))
```

```python
#------------------Exercice---------------------
dataset = {f'experience{i}' : np.random.randn(100) for i in range(4)  
}
```

```python
def graphique(dataset):
    n= len(dataset)
    plt.figure(figsize=(10,8))
    for k, i in zip(dataset.keys(), range(1, n+1)):
        plt.subplot(n,1,i)
        plt.plot(dataset[k])
        plt.title(k)
    plt.show()
    

graphique(dataset)
        
```
