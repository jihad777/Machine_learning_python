```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2,8)
y = x**2
plt.scatter(x,y)
```

```python
#---------------Interpolation----------------------
from scipy.interpolate import interp1d
```

```python
f = interp1d(x,y, kind='linear') #kind = cubique, quadratic...

new_x = np.linspace(0,2,30)
new_y= f(new_x)

plt.scatter(x,y)
plt.scatter(new_x, new_y, c='pink')
```

```python
#----------------Optimization (with scipy) >> not commonly used, preffer (scikit-learn)----------------------
#excample
def f(x, a, b, c, d):
    return a*x**3 + b*x**2 + c*x + d
```

```python
from scipy import optimize
```

```python
params , params_covariance = optimize.curve_fit(f, x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, f(x, params[0], params[1], params[2], params[3]))
```

```python
#-------------------------Minimisation-------------------------
def f(x):
    return x**2 +15*np.sin(x)
```

```python
x = np.linspace(-10,10,100)
plt.plot(x, f(x))
```

```python
result = optimize.minimize(f, x0=-3).x
plt.plot(x, f(x))
plt.scatter(result, f(result),s = 100, marker = 'x',  c='r', zorder = 1) #s = size of dot
```
